## Data visualization

## Which are the most comercial model ?

```python
most_registered_model = df['Model'].value_counts().nlargest(10).reset_index()
most_registered_model
```

```python
most_registered_model_colums = ['Model', 'count']
sns.barplot(x='Model', y='count', data = most_registered_model)
plt.title('Most registered Model in Washington')
plt.xlabel('Electric Vehicel model')
plt.ylabel('Quantity Registered')
plt.xticks(rotation=90)
```

## Report
In this graph, we can see the different models of electric cars registered in the state of Washington. It is evident that the models predominantly represented in the graph belong to Tesla. It can be said that Tesla is the company that has made the most advancements in the field of electric cars, also implementing autonomous vehicles.

The other models are from various brands such as Nissan, Chevrolet, Ford, etc. While they may not have as much impact as Tesla, they still hold a certain percentage in the graph. These models are likely more affordable, but perhaps not the first choice for individuals looking to purchase from these brands.

A significant factor in why Tesla also dominates this percentage is that all Tesla vehicles are electric. Therefore, when collecting this type of information, it is inevitable for Tesla to appear in these situations.s  


## Which are the cities with most electric vehicles ?

```python
cities_most_electric_vehicle = df['City'].value_counts().nlargest(10).reset_index()
cities_most_electric_vehicle
```

## Which are the brands with the most electric vehicles ?


```python
most_registered_model_colums = ['Model', 'count']
sns.barplot(x='City', y='count', data = cities_most_electric_vehicle)
plt.title('Cities with most registered Electric Vehicles')
plt.xlabel('City')
plt.ylabel('Quantity Registered')
plt.xticks(rotation=90)
```

## Report
As we can see in the following graph, we can observe how electric vehicles are distributed in the state of Washington in different cities. Clearly, Seattle stands out with a significantly notable percentage of electric vehicles compared to the others.
This graph allows us to compare the number of registered vehicles in each city. The graph indicates that, unlike other cities, Seattle is taking the lead in implementing electric vehicles for daily use, as gasoline vehicles contribute significantly to environmental pollution, while electric vehicles help mitigate this environmental impact.
Another factor that likely contributes to this significant difference with Seattle is probably the purchasing power of other cities. As mentioned earlier, Tesla models are the most dominant, being more modern than their competitors, but this does not change the fact that Tesla cars are quite expensive and not everyone has the financial capacity to acquire this type of vehicle.
To conclude, it can be said that electric cars are a relatively new technology. In this graph, the percentage of people who still prefer gasoline vehicles is not ruled out, as these vehicles can be more durable and reliable. Although the graph shows that electric cars are more dominant in Seattle, it does not mean that electric vehicles prevail over gasoline vehicles in Seattle.

```python
electrics_brands = df['Make'].value_counts().nlargest(5).reset_index()
electrics_brands
```

```python
labels = ['Tesla' , 'Nissan', 'chevrolet', 'Ford', 'BMW']
plt.pie(electrics_brands['count'], autopct = '%1.1f%%', labels= labels )
plt.title('Brands with Most Electric Vehicles')
plt.show()
```

## Report

In the following graph, we can observe the distribution of different electric vehicle brands in the state of Washington. As mentioned earlier, it is no coincidence that Tesla dominates as the most predominant electric vehicle brand.

This is due to Tesla's vehicles having autonomous driving capabilities, meaning they can drive themselves. While Tesla vehicles are not cheap, the various features they offer justify their price.

On the other hand, other brands do not hold as significant a percentage, posing no real challenge to Tesla in the market. These brands need to identify what Tesla is doing differently in order to capture a larger share and increase competitiveness.
